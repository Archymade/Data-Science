```python
from sklearn.preprocessing import LabelEncoder, Binarizer
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler

from sklearn.feature_selection import SelectKBest, mutual_info_classif
from sklearn.preprocessing import PowerTransformer
from sklearn.compose import ColumnTransformer

from sklearn.pipeline import Pipeline
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV

from sklearn.model_selection import cross_val_score, RepeatedStratifiedKFold
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor
```

```python
sel = SelectKBest(score_func = mutual_info_classif, k = 9)
```

```python
labelEncoder = LabelEncoder()
binarizer = Binarizer(threshold = 3)

y = binarizer.fit_transform(y)
y = labelEncoder.fit_transform(y)
```

```python
RFC = RandomForestClassifier(random_state = 42)
```

```python
norm = MinMaxScaler()
scaler = RobustScaler()
power = PowerTransformer()
```

```python
t = [('power', power, [0, 1, 3, 4, 5, 6, 7, 8, 9, 10]),
     ('scaler', scaler, [2])]

colTrans = ColumnTransformer(transformers = t, remainder = 'passthrough')
```

```python
cv = RepeatedStratifiedKFold(n_splits = 3, n_repeats = 5, random_state = 42)
```

### Preparing the final estimator pipeline with the determined optimal hyperparameters...

```python
RFC = RandomForestClassifier(criterion = 'gini',
                             max_depth = 7,
                             n_estimators = 250)

norm = MinMaxScaler()

power = PowerTransformer(method = 'yeo-johnson')

# Since the optimal k-value for the SelctKBest class is 'all', there is no need to use the class.
```

```python
final_pipeline = Pipeline(steps = [('scaler', scaler),
                                   ('norm', norm),
                                   ('power', power),
                                   ('RFC', RFC)])
```

```python
scoresA = cross_val_score(estimator = final_pipeline, X = X, y = y, cv = cv, scoring = 'roc_auc_ovr', n_jobs = -1)
```

```python
def results(scores):
    
    ''' Prints out the average roc_auc_ovr score and the mean standard deviation. '''
    
    print(f'Mean roc_auc_ovr score : {np.mean(scores)}')
    print('----------')
    print(f'Mean standard deviation score (std) : {np.std(scores)}')
```

```python
results(scoresA)
```

### Using ColumnTransformer...

```python
param_net = {'colTrans__power__method' : ['box-cox', 'yeo-johnson'],
             'colTrans__scaler__with_centering' : [True, False],
             'RFC__n_estimators' : [100, 150, 200, 250, 300],
             'RFC__criterion' : ['gini', 'entropy'],
             'sel__k' : [7, 8, 9, 10, 'all'],
             'RFC__max_depth' : [3, 4, 5, 6, 7]}
```

```python
RFC = RandomForestClassifier(n_estimators = 250,
                             max_depth = 7,
                             criterion = 'gini',
                             random_state = 42)

power = PowerTransformer(method = 'box-cox')
scaler = RobustScaler(with_centering = True)

t = [('power', power, [0, 1, 3, 4, 5, 6, 7, 8, 9, 10]),
     ('scaler', scaler, [2])]

colTrans = ColumnTransformer(transformers = t, remainder = 'passthrough')

# No need for SelectKBest class, as we will be using all features.
```

```python
final_pipelineB = Pipeline(steps = [('colTrans', colTrans),
                                    ('norm', norm),
                                    ('RFC', RFC)])
```

```python
scoresB = cross_val_score(estimator = final_pipelineB, X = X, y = y, cv = cv, scoring = 'roc_auc_ovr', n_jobs = -1)
```

```python
def resultsB(scores):
    
    ''' Prints out the average roc_auc_ovr score and the mean standard deviation. '''
    
    print(f'Mean roc_auc_ovr score : {np.mean(scoresB)}')
    print('----------')
    print(f'Mean standard deviation score (std) : {np.std(scoresB)}')
```

```python
resultsB(scoresB)
```

### From the results above, it is evident that applying select transforms to specific columns of data is more effective performance-wise than the general transform application. This makes sense as some columns may require some unique tranformation requirements.

```python
def final_model():
    final_pipelineB = Pipeline(steps = [('colTrans', colTrans),
                                        ('norm', norm),
                                        ('RFC', RFC)])
    
    final_model = final_pipelineB.fit(X, y)
    
    scores = cross_val_score(estimator = final_model, X = X, y = y, cv = cv, scoring = 'roc_auc_ovr', n_jobs = -1)
    
    return resultsB(scores)
```
